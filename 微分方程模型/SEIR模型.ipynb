# SEIR模型

![SEIR模型](picture/SEIR.png)

```python
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SEIR_model(y, t, alpha, lamb, beta):
    S, E, I, R = y
    dS_dt = -alpha * (E + I) * S
    dE_dt = alpha * (E + I) * S - lamb * E
    dI_dt = lamd * E - beta * I
    dR_dt = beta * I
    return ([dS_dt, dE_dt, dI_dt, dR_dt])
```

```python
t = np.linspace(0, 100, 10000)
S0 = 0.8
E0 = 0.1
I0 = 0.1
R0 = 0.0
alpha = 0.35
lamd = 0.2
beta = 0.1
res = scipy.integrate.odeint(SEIR_model,[S0, E0, I0, R0], t, args=(alpha, lamd, beta))
```

```python
plt.figure(figsize=(12, 8))
plt.plot(res[:, 0], t, label='S(t)')
plt.plot(res[:, 1], t, label='E(t)')
plt.plot(res[:, 2], t, label='I(t)')
plt.plot(res[:, 3], t, label='R(t)')
plt.grid()
plt.legend()
plt.show()
```
