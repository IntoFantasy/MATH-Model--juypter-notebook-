# 生物竞争模型

![模型选择](picture/模型选择.png)

![向量场](picture/向量场1.png)

```python
import numpy as np
import matplotlib.pyplot as plt

a = 2
b = 1
c = 2
d = 1

X, Y = np.meshgrid(np.linspace(0, 3, 10), np.linspace(0, 3, 10))
u = (a - b * Y) * X
v = (c - d * X) * Y
plt.figure(figsize=(12, 8))
plt.quiver(X, Y, u, v, color='red')
plt.show()
```

```python
plt.streamplot(X, Y, u, v, color='red')
plt.show()
```

# 捕食模型

```python
u = (a - b * Y) * X
v = (d * X - c) * Y
plt.figure(figsize=(12, 8))
plt.quiver(X, Y, u, v, color='green')
plt.streamplot(X, Y, u, v, color='blue')
plt.show()
```
