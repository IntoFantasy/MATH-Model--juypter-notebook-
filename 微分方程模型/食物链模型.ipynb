# 食物链

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 三者竞争

![三者竞争](picture/三者竞争.png)

```python
a1, b1, a2, b2, a3, b3 = 2, 1, 2, 1, 2, 1
def model(y, t, a1, b1, a2, b2, a3, b3):
    X, Y, Z = y
    dx = (a2 - b2 * Y) * (a3 - b3 * Z) * X
    dy = (a1 - b1 * X) * (a3 - b3 * Z) * Y
    dz = (a2 - b2 * Y) * (a1 - b1 * X) * Z
    return ([dx, dy, dz])
import scipy.integrate as scp
t = np.linspace(0, 500, 10000)
x0, y0, z0 = 0.3, 0.5, 0.7
res = scp.odeint(model, [x0, y0, z0], t, args=(a1, b1, a2, b2, a3, b3))

plt.figure(figsize=(12, 8))
ax = plt.axes(projection='3d')
ax.plot3D(res[:, 0], res[:, 1], res[:, 2], 'red')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(30, 15)
plt.show()
```

### A, B捕食C

![A,B捕食C](picture/A，B捕食C.png)

```python
def model_2(y, t, a1, b1, a2, b2, a3, b3):
    X, Y, Z = y
    dx = -(a2 - b2 * Y) * (a3 - b3 * Z) * X
    dy = -(a1 - b1 * X) * (a3 - b3 * Z) * Y
    dz = (a2 - b2 * Y) * (a1 - b1 * X) * Z
    return ([dx, dy, dz])
import scipy.integrate as scp
t = np.linspace(0, 10000, 1000000)
x0, y0, z0 = 0.3, 0.5, 0.7
res = scp.odeint(model_2, [x0, y0, z0], t, args=(a1, b1, a2, b2, a3, b3))
plt.figure(figsize=(24, 16))
ax = plt.axes(projection='3d')
ax.plot3D(res[:, 0], res[:, 1], res[:, 2], 'red')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter3D(res[:, 0], res[:, 1], res[:, 2], 'green')
ax.view_init(20, 45)
plt.show()
```
